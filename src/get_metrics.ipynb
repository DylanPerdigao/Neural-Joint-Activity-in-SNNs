```python
import pandas as pd
import sys
sys.path.append('../../')
```

```python
configs = [
    "P2_S20",
    "P20_S20",
    "P200_S20",
    "P2_S50",
    "P20_S50",
    "P200_S50"
]
dfs = []
for config in configs:
    df = pd.read_csv(f'results/results-{config}.csv')
    df['config'] = config
    dfs.append(df)
df = pd.concat(dfs)
df = df[df['recall@5FPR'] > 0]
df_aux = df.groupby(['config','dataset']).agg({
    'recall@5FPR': ['mean', 'std'],
    'fpr@5FPR': ['mean', 'std'],
    'fpr_ratio_age': ['mean', 'std'],
    'fpr_ratio_income': ['mean', 'std'],
    'fpr_ratio_employment': ['mean', 'std'],
})
df_aux = df_aux * 100
df_aux
```
