```python
import optuna
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
configs = [
    'S20_P2',
    'S20_P20',
    'S20_P200',
    'S50_P2',
    'S50_P20',
    'S50_P200', 
]
best_trials = pd.DataFrame()
for config in configs:
    study = optuna.load_study(study_name=config, storage='sqlite:///../ciarp2024.db')
    df = study.trials_dataframe()
    df = df[df['values_0'] <= 0.05]
    df = df.sort_values(by='user_attrs_@5FPR recall', ascending=False)
    df = df.head(1)
    df['population'] = int(config.split('P')[1].split('_')[0])
    df['steps'] = int(config.split('S')[1].split('_')[0])
    best_trials = pd.concat([best_trials, df])
best_trials = best_trials.sort_values(by='user_attrs_@5FPR recall', ascending=False)
best_trials_selected = best_trials[[
    'steps', 'population',
    'params_beta1','params_beta2','params_beta3','params_beta4',
    'params_slope',	
    'params_threshold1','params_threshold2','params_threshold3', 'params_threshold4',
    'params_weight',
    'params_adam_beta1','params_adam_beta2',
    'params_learning_rate',	
    'user_attrs_@5FPR recall',
    'user_attrs_@5FPR fpr',
]]
best_trials_selected_sorted = best_trials_selected.sort_values(by=['steps', 'population'], ascending=True).T
best_trials_selected_sorted.columns = configs
best_trials_selected_sorted
```
